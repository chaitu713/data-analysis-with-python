Step 4 : Cleaning the Data

- Dropping the show_id column as we are not going to use that in our analysis

```python
df.drop(['show_id'], axis = 1, inplace = True)
```

```python
df.head(10)
```

- Checking for duplicates records of TV Shows/ Movies and deleting duplicates

```python
df.drop_duplicates(["title"])
df.shape
```

- Replacing null values in 'cast', 'date_added' and 'country' with 'Unavailable'

```python
df['country'].fillna('Unavailable', inplace = True)
df['date_added'].fillna('unavailable', inplace = True)
df['cast'].fillna('Unavailable', inplace = True)
```

- Replacing missing values in 'rating' with mode of the column

```python
df["rating"].fillna(df['rating'].mode()[0], inplace = True)
df['director'].fillna(df['director'].mode()[0], inplace = True)
```

- Renaming 'listed_in' column to 'genre' for better understanding

```python
df = df.rename(columns = {"listed_in" : "genre"})
df['genre'] = df['genre'].apply(lambda x : x.split(",")[0])
df.head(10)
```

## Exploratory Data Analysis and Visualization

- Analysing entire Prime Video dataset of both movies and shows

Let's begin by importing`matplotlib.pyplot` and `seaborn`.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

Question 1 : Distribution of type of data in the Prime Video dataset

```python
group_type = df["type"].value_counts().reset_index()
group_type = group_type.rename(columns = {"index": "type", "type" : "count"})
group_type
```

```python
plt.figure(figsize = (12,6))
colors = sns.color_palette('pastel')[0:2]
plt.title("Percentage of titles that are Movie / TV Shows")
plt.pie(group_type['count'], labels = group_type['type'], colors = colors, autopct='%.0f%%');
```

- So, our Prime Video dataset contains of 
  - Movies with count - 7814 and percentage distribution - 81% 
  - TV Shows with count - 1854 and percentage distribution - 19%

Question 2 : Distribution of Ratings in the dataset

```python
df["rating"].unique()
```

- We can combine the some ratings category to one single category 
  - For Example:  NOT_RATE and UNRATED can be combined to say UNRATED

```python
for i in df.index:
  if(df.loc[i,'rating'] == 'NOT_RATE'):
    df[i,'rating'] = "UNRATED"
  elif(df.loc[i,'rating'] == 'AGES_18_'):
    df[i,'rating'] = "18+"
  elif((df.loc[i,'rating'] == 'AGES_16_') or (df.loc[i,'rating'] == '16')):
    df[i,'rating'] = "16+"
  elif(df.loc[i,'rating'] == 'ALL_AGES'):
    df[i,'rating'] = 'G'
  else:
    continue
```

```python
plt.figure(figsize = (25,10))
sns.countplot(x = "rating", hue = "type", data=df, palette = "Set2");
plt.title("Distribution of Prime Movies & TV Shows across ratings");
plt.legend(loc = 'upper center', title = 'Type', fontsize = 16);
```

Question 3 : What is the most popular rating category

```python
popular_rating = df["rating"].value_counts()
```

```python
plt.figure(figsize = (18,10))
popular_rating.plot(color = 'green', marker = 'o', markerfacecolor = 'red');
```

Question 4 : What is the Distribution of Prime Movies & TV Shows across genres

```python
group_genre = df.groupby('type')["genre"].value_counts()
group_genre
```

## Asking and Answering Questions

Let's ask some specific questions, and try to answer them using dataset operations and interesting visualizations.



#### Q1: Which actors films are more in Prime Video?

```python
plt.figure(figsize=(15,9))
plt.title('Top 10 Actors Based on The Number of Titles')
filtered_cast = df[df.cast != 'Unavailable'].set_index('title').cast.str.split(', ', expand=True).stack().reset_index(level=1, drop=True)
sns.countplot(y = filtered_cast, order=filtered_cast.value_counts().index[:10], palette='pastel')
plt.show()
```

- Actor with more number of movies in Prime Video is Maggie Binkley

#### Q2: Which directors movies are more in Prime Video?

```python
filtered_directors = df[df.director != 'Unavailable'].set_index('title').director.str.split(', ', expand=True).stack().reset_index(level=1, drop=True)
plt.figure(figsize=(15,9))
plt.title('Top 10 Director Based on The Number of Titles')
sns.countplot(y = filtered_directors, order=filtered_directors.value_counts().index[:5], palette='Set1')
plt.show()
```

- Director with most of films in Prime Video is Mark Knight

#### Q3: TODO - Which 15 countries are top contributors to Prime Video

```python
filtered_countries = df.set_index('title').country.str.split(', ', expand=True).stack().reset_index(level=1, drop=True);
filtered_countries = filtered_countries[filtered_countries != 'Unavailable']
plt.figure(figsize=(15,9))
g = sns.countplot(y = filtered_countries, order=filtered_countries.value_counts().index[:15], palette = 'Set2')
plt.title('Top 15 Countries Contributor on Netflix')
plt.xlabel('Titles')
plt.ylabel('Country')
plt.show()
```

- The top contributor to Prime Video is United States

#### Q4: Number of Movies/TV Shows released per year after 2010

```python
plt.figure(figsize = (25,15))
sns.countplot(y = "release_year", hue = "type", data=df[df['release_year'] > 2010], palette = "Wistia");
plt.title("Distribution of Prime Movies & TV Shows across released years after 2010");
plt.legend(loc = 'upper right', title = 'Type', fontsize = 20);
```

- So more number of Movies/ TV shows released in year 2021

#### Q5: TODO - Number of Movies/TV Shows released per year before 1950
